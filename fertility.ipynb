```python
import kagglehub
import pandas as pd
import os

# Download dataset
path = kagglehub.dataset_download(
    "muhamedyoussry/fertility-data-set"
)

# Show available files
print("Files in dataset:", os.listdir(path))

# Load the CSV file (replace the filename if needed)
file_name = os.listdir(path)[0]

df = pd.read_csv(os.path.join(path, file_name))
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
set(df['Frequency of alcohol consumption'])
```

```python
d={'every day':7,
 'hardly ever or never':0,
 'once a week':1,
 'several times a day':14,
 'several times a week':4}
```

```python
df['Frequency of alcohol consumption']=df['Frequency of alcohol consumption'].map(d)
```

```python
d={'yes':1,'no':0}
```

```python
df['Childish diseases']=df['Childish diseases'].map(d)
```

```python
df['Accident or serious trauma']=df['Accident or serious trauma'].map(d)
```

```python
df['Surgical intervention']=df['Surgical intervention'].map(d)
```

```python
set(df['Smoking habit'])
```

```python
d={'daily':7, 'never':0, 'occasional':3}
```

```python
df['Smoking habit']=df['Smoking habit'].map(d)
```

```python
set(df['High fevers in the last year'])
```

```python
d={'less than 3 months ago':1, 'more than 3 months ago':2, 'no':0}
```

```python
df['High fevers in the last year']=df['High fevers in the last year'].map(d)
```

```python
df['Season'].unique()
```

```python
import numpy as np

season_map = {
    "spring": 0,
    "summer": 1,
    "fall": 2,
    "winter": 3
}

x = df["Season"].map(season_map)

df["season_sin"] = np.sin(2 * np.pi * x / 4)
df["season_cos"] = np.cos(2 * np.pi * x / 4)
```

```python
df.drop('Season',axis=1,inplace=True)
```

```python
y=df['Diagnosis']
```

```python
X=df.drop('Diagnosis',axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=90)
```

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
model=KNeighborsClassifier(n_neighbors=9)
model.fit(x_train,y_train)
ypred=model.predict(x_test)
accuracy_score(ypred,y_test)
```
