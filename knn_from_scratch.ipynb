```python
import numpy as np


class knn():
    def __init__(self,k=3):
        self.k=k
    def fit(self,X,y):
        self.X=X
        self.y=y
    def distance(self,v1,v2):
        return sum((v1-v2)**2)**0.5
    def predict_one(self,x):
        ypred=[self.distance(x,i) for i in self.X]
        ypred=np.argsort(ypred)[:self.k]
        yp=[self.y[i] for i in ypred]
        yp=np.bincount(ypred).argmax()
        return yp
    def predict(self,X):
        ypred=[self.predict_one(x) for x in X]
        return ypred
       
        
   
        


        
```
