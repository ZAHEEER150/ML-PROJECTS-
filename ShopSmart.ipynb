```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Zaheer\Downloads\shop_smart_ecommerce.csv")
```

```python
df=df.drop_duplicates()
df.dtypes
df.head()
```

```python
df=pd.get_dummies(df,columns=['VisitorType'],drop_first=True,dtype=int)
```

```python
bool_cols = df.select_dtypes(include='bool').columns

df[bool_cols] = df[bool_cols].astype(int)
```

```python
df['Month']
```

```python
import numpy as np

month_map = {
    'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4,
    'May':5, 'Jun':6, 'Jul':7, 'Aug':8,
    'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12
}

df['Month_num'] = df['Month'].map(month_map)

df['Month_sin'] = np.sin(2*np.pi*df['Month_num']/12)
df['Month_cos'] = np.cos(2*np.pi*df['Month_num']/12)

df = df.drop(columns=['Month', 'Month_num'])
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.model_selection import train_test_split
```

```python
y=df['Revenue']
X=df.drop('Revenue',axis=1)
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=88)

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
```

```python
model=DecisionTreeClassifier(min_samples_leaf=45)
model.fit(x_train,y_train)
ypred=model.predict(x_test)
f1_score(y_test, ypred)
```
